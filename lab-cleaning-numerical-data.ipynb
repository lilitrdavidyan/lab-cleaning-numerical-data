### 7: Differentiate between continuous and discrete variables

```python
numerical_df.nunique()
```

```python
numerical_df.describe().T
```

```python
def differentiate_numerical(df, threshold=40):
    continuous_vars = []
    discrete_vars = []
    
    for column in df.columns:
        unique_values = df[column].nunique()
        if unique_values > threshold:
            continuous_vars.append(column)
        else:
            discrete_vars.append(column)
            
    return continuous_vars, discrete_vars
```

```python
continuous_vars, discrete_vars = differentiate_numerical(numerical_df)
```

```python
continuous_vars
```

### 9: Plot discrete and continuous variables

```python
def plot_discrete_variables(dataframe, variables):
    for var in variables:
        plt.figure(figsize=(10, 3))
        sns.countplot(data=dataframe, x=var)
        plt.title(f'Distribution of {var}')
        plt.show()

def plot_continuous_variables(dataframe, variables):
    for var in variables:
        plt.figure(figsize=(10, 3))
        sns.histplot(data=dataframe, x=var, kde=True, bins=50)
        plt.title(f'Distribution of {var}')
        plt.show()
```

### 11: Look for outliers in the continuous variables

```python
def plot_outliers(dataframe, variables):
    for var in variables:
        plt.figure(figsize=(10, 3))
        sns.boxplot(data=dataframe, x=var)
        plt.title(f'Boxplot of {var}')
        plt.show()
```

- **Total Claim Amount**: Several points lie outside the whiskers, indicating potential outliers.

### 12: Commenting and dealing with the outliers

```python
def remove_outliers(dataframe, variables):
    df = dataframe.copy()
    for var in variables:
        Q1 = df[var].quantile(0.25)
        Q3 = df[var].quantile(0.75)
        IQR = Q3 - Q1
        
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        df = df[(df[var] >= lower_bound) & (df[var] <= upper_bound)]
    return df
```

```python
vars_with_outliers = ['customer_lifetime_value', 'monthly_premium_auto', 'total_claim_amount']
```

```python
cleaned_numerical_df = remove_outliers(numerical_df, vars_with_outliers)
```

```python
numerical_df.shape
```

```python
cleaned_numerical_df.shape
```

### Observations: 
- Loss of Information: The removal of 1,345 rows is about 15% of the original dataset. Removing such a considerable portion of data might lead to a loss of important information.
- Potential Bias: If outliers were systematically removed, it could introduce bias into the dataset. For instance, if the outliers represent high-value customers, excluding them would skew the analysis.
- Generalization: Removing outliers can improve the performance of certain models on the training dataset, but it might lead to a model that generalizes poorly to new, unseen data.


**Transformation**: Instead of removing outliers, we can apply log transformations that can reduce the impact of extreme values.

The logarithm of a non-positive number is undefined. Hence, we can only apply this transformation to columns with strictly positive values.

```python
def log_transform(dataframe, variables):
    df = dataframe.copy()
    for var in variables:
        # Add a small constant to avoid log(0)
        df[var] = np.log(df[var] + 1)
    return df
```

```python
log_transformed_df = log_transform(numerical_df, vars_with_outliers)
log_transformed_df
```

```python
plot_continuous_variables(log_transformed_df, vars_with_outliers)
```

```python
plot_outliers(log_transformed_df, vars_with_outliers)
```

### Observations from the Plots of Log-transformed Columns:

1. **Customer Lifetime Value**:
    - **Histogram**: The distribution appears bimodal, indicating two distinct peaks. This might suggest the existence of two groups or segments within this data.
    - **Boxplot**: The distribution is more compact compared to its original state. There are still some data points beyond the upper whisker, which could be considered mild outliers, but they are less extreme than before.

2. **Monthly Premium Auto**:
    - **Histogram**: While still slightly right-skewed, the distribution is more symmetrical after the log transformation and is closer to being normally distributed.
    - **Boxplot**: The range of values is tighter, and although there remain potential outliers, they are not as pronounced as in the untransformed data.

3. **Total Claim Amount**:
    - **Histogram**: The distribution now appears more symmetrical and much closer to a normal distribution.
    - **Boxplot**: The values are well-distributed with a few potential outliers, but these are far less extreme than previously observed.

**General Observations**:
- The log transformation has made the data distributions of these columns more compact and, in some cases, more symmetrical. This will likely reduce the impact of extreme values in analytical or modeling tasks.
- Some potential outliers remain post-transformation, but these are less severe than those in the original data. Given the nature of the transformed data, these "outliers" might not be as influential or problematic as their original counterparts.
